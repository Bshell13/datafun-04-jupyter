```python
import matplotlib as plt
import pandas as pd
import seaborn as sns

# pd.read_csv will read a csv file using pandas
df = sns.load_dataset("penguins") # load dataset into a DataFrame
```

```python
print(df.head(10)) # Displays the top 10 penguins in the dataset by index
print(df.shape) # Displays the matrix indecies
print(df.dtypes) # Displays the column data types
```

```python
print(df.describe()) # calculates the basics statistics for each column
```

```python
# .hist() builds a histogram
df['bill_length_mm'].hist() # one column
df.hist() # all columns
```

```python
# .value_counts() displays the count of each category
df['species'].value_counts() # one column
```

```python
# renaming columns
df.rename(columns={'species': 'Penguin Species'}, inplace= True)

# adding columns to dataset
df['body_mas_lbs'] = df['body_mass_g'] * 0.00220462
```

```python
# Creating plots using seaborn and matplotlib

# Creating a scatter plot of flipper length (mm) and botdy mass (g)
plt = sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="Penguin Species")

# Setting axis labels and title
plt.set_xlabel("Flipper length (mm)")
plt.set_ylabel("Body mass (g)")
plt.set_title("Penguin Flipper Length vs. Body Mass")

```
